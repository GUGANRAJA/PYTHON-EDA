# FEATURE ENGINEERING

```python
df['Date_of_Journey'].str.split('/').str[0]
```

```python
#date we have to convert into date format.
df['day']=df['Date_of_Journey'].str.split('/').str[0]
df['month']=df['Date_of_Journey'].str.split('/').str[1]
df['year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
#lambda also used to get same output
#df['day']=df['Date_of_Journey'].apply(lambda x:x.split("/")[0])
#df['month']=df['Date_of_Journey'].apply(lambda x:x.split("/")[1])
#df['year']=df['Date_of_Journey'].apply(lambda x:x.split("/")[2])
```

```python
df.head()
```

```python
df['day']=df['day'].astype(int)
df['month']=df['month'].astype(int)
df['year']=df['year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df['Arrival_Time'].str.split(' ').str[0] #lambda also used.
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(" ")[0])
```

```python
df.isnull().sum()
```

```python
df['Arrival_hour']=df['Arrival_Time'].apply(lambda x:x.split(":")[0])
df['Arrival_min']=df['Arrival_Time'].apply(lambda x:x.split(":")[1])
```

```python
df['Arrival_min']=df['Arrival_min'].astype(int)
df['Arrival_hour']=df['Arrival_hour'].astype(int)
```

```python
df.drop("Arrival_Time", axis=1, inplace=True)
```
